```python
# Problem 1: Smart Parking Lot Management System
def parking_system(capacity, logs):
    current = 0
    peak = 0
    for log in logs:
        if log == "IN":
            current += 1
        elif log == "OUT" and current > 0:
            current -= 1
        if current > peak:
            peak = current
    status = "Full" if current >= capacity else "Available"
    print("Currently Parked Vehicles:", current)
    print("Parking Status:", status)

parking_system(50, ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"])
```

```python
# Problem 2: Online Food Delivery Time Estimator
def estimate_delivery(distance, traffic, weather):
    base_time = distance * 5
    if traffic == "High":
        base_time += 15
    elif traffic == "Medium":
        base_time += 10
    if weather == "Rainy":
        base_time += 10
    print("Estimated Delivery Time:", base_time, "minutes")

estimate_delivery(8, "High", "Rainy")
```

```python
# Problem 3: Movie Theatre Seat Occupancy Analyzer
def seat_occupancy(total_seats, booked_seats):
    occupancy = (len(booked_seats) / total_seats) * 100
    if occupancy == 100:
        status = "Housefull"
    elif occupancy >= 75:
        status = "Almost Full"
    else:
        status = "Seats Available"
    print("Occupancy:", round(occupancy), "%")
    print("Show Status:", status)

seat_occupancy(200, [1]*150)
```

```python
# Problem 4: Cloud Server Load Classification System
def server_load(cpu_readings):
    total = 0
    for reading in cpu_readings:
        total += reading
    avg = total / len(cpu_readings)
    if avg < 50:
        status = "Normal"
    elif avg <= 80:
        status = "Warning"
    else:
        status = "Critical"
    print("Average CPU Load:", round(avg), "%")
    print("Server Status:", status)

server_load([45, 60, 70, 85, 90])
```

```python
# Problem 5: Smart Classroom Resource Usage Monitor
def resource_monitor(resources):
    overused = []
    for resource, hours in resources.items():
        if hours > 8:
            overused.append(resource)
    alert = "Yes" if overused else "No"
    print("Overused Resources:", ", ".join(overused))
    print("Energy Alert:", alert)

resource_monitor({"Projector": 6, "AC": 9, "Lights": 4})
```

```python
# Problem 6: Online Event Registration Capacity Controller
def registration_system(capacity, registrations):
    confirmed = min(capacity, registrations)
    waitlisted = max(0, registrations - capacity)
    status = "Closed" if registrations >= capacity else "Open"
    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlisted)
    print("Registration Status:", status)

registration_system(100, 105)
```
